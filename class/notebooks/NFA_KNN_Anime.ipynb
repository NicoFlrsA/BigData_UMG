```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/AnimeList.csv")
df
```

```python
x = df[["episodes","rating"]]
x.info()
```

```python
x["episodes"] = x["episodes"].replace("Unknown",1)
x["episodes"] = x["episodes"].astype("int32")
set(x["episodes"].values)
```

```python
x.info()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x["episodes"],x["rating"])
#plt.scatter(x[:, 0],x[:, 1])
```

```python
from sklearn.neighbors import NearestNeighbors

x_cleaned = x.dropna()  
model = NearestNeighbors(n_neighbors=2, metric = "minkowski").fit(x_cleaned)
```

```python
model
```

```python
test = [[12, 10]]

model.kneighbors(test)
```
