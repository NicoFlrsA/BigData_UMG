```python
df.drop(["Poster_Link", "Overview", "Runtime", "Series_Title", "Gross", "Certificate"], axis=1, inplace=True)
df
```

```python

df.fillna(df.mode, inplace=True)
```

```python
pd_processed = pd.get_dummies(df)
pd_processed.info()
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=3, metric="cosine")
model.fit(pd_processed)
```
